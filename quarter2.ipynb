```python
January = pd.read_csv("Sales_January_2019.csv")
February = pd.read_csv("Sales_February_2019.csv")
March = pd.read_csv("Sales_March_2019.csv")
April = pd.read_csv("Sales_April_2019.csv")
May = pd.read_csv("Sales_May_2019.csv")
June = pd.read_csv("Sales_June_2019.csv")
July = pd.read_csv("Sales_July_2019.csv")
August = pd.read_csv("Sales_August_2019.csv")
September = pd.read_csv("Sales_September_2019.csv")
October = pd.read_csv("Sales_October_2019.csv")
November = pd.read_csv("Sales_November_2019.csv")
December = pd.read_csv("Sales_December_2019.csv")
```

```python
aprilmay = April.append(May)
qt2 = aprilmay.append(June)
```

```python
filter = qt2['Quantity Ordered'] != 'Quantity Ordered'
qt2 = qt2[filter]
qt2 = qt2.dropna()
qt2
## To eliminate any instances of the headers showing up in the data
```

```python
qt2[['Order Month', 'Order Time']] = qt2["Order Date"].str.split(' ', 1, expand=True)

qt2[['Address', 'City/State/ZIP']] = qt2["Purchase Address"].str.split(',', 1, expand=True)

qt2[['Month', 'Day', 'Year']] = qt2['Order Month'].str.split("/", 2, expand=True)

qt2
```

```python
qt2['Quantity Ordered'] = pd.to_numeric(qt2['Quantity Ordered'])
qt2['Price Each'] = pd.to_numeric(qt2['Price Each'])
qt2['Month'] = pd.to_numeric(qt2['Month'])
qt2['Day'] = pd.to_numeric(qt2['Day'])
qt2['Year'] = pd.to_numeric(qt2['Year'])
qt2
```

```python
qt2 = qt2.drop(columns=["Order Month", "Purchase Address", "Order Date"])
qt2 = qt2[qt2['Month'] <= 6]
qt2
```

```python
product_count = qt2.groupby('Month').count()
product_count
```

```python
plt.pie(product_count["Quantity Ordered"], labels=['April', 'May', 'June'], colors=['red', 'pink', 'steelblue'], autopct='%1.0f%%')
plt.show()
```

```python
region = qt2.groupby('City/State/ZIP').count()[['Quantity Ordered']]
region
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=region.reset_index(), x='City/State/ZIP', y='Quantity Ordered')
plt.xticks(rotation=90)
plt.show()
```

```python
monthday = qt2.groupby(['Month', 'Day']).count()[['Quantity Ordered']]
monthday
```

```python
product_price = qt2[['Product','Price Each']].drop_duplicates(subset=['Product','Price Each'])
product_price
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(
    data=product_price,
    x = 'Product',
    y = 'Price Each'
)
plt.xticks(rotation=90)
plt.show()
```

```python
sales_by_state = qt2.groupby(['City/State/ZIP','Product']).count()
sales_by_state
```

```python
sns.set(rc={'figure.figsize':(20,15)})
sns.barplot(
    data=sales_by_state.reset_index(),
    x = 'City/State/ZIP',
    y = 'Order ID',
    hue = 'Product'
)
plt.xticks(rotation=-45)
```
