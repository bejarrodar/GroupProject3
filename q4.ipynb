```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
October = pd.read_csv('Sales_October_2019.csv')
November = pd.read_csv('Sales_November_2019.csv')
December = pd.read_csv('Sales_December_2019.csv')
```

```python
q4 = pd.concat([October, November, December])
q4
```

```python
filter = q4['Quantity Ordered'] != 'Quantity Ordered' 
q4= q4[filter]
q4
```

```python
q4 = q4.dropna()
q4
```

```python
q4.head()
```

```python
sns.heatmap(data=pd.isnull(q4))
```

```python
q4['city'] = q4['Purchase Address'].str.split(", ",expand=True)[1]
q4
```

```python
q4['state'] = q4['Purchase Address'].str.split(", ",expand=True)[2].str.split(" ", expand=True)[0]
q4
```

```python
q4['month'] = q4['Order Date'].str[0:2]
q4['year'] = q4['Order Date'].str[6:8]
q4['time'] = q4['Order Date'].str[-6:-3]
q4
```

```python
filter = q4['month'] != 1
q4 = q4[filter]
```

```python
q4['Quantity Ordered'] = pd.to_numeric(q4['Quantity Ordered'])
q4['Price Each'] = pd.to_numeric(q4['Price Each'])
q4['month'] = pd.to_numeric(q4['month'])
q4['year'] = pd.to_numeric(q4['year'])
q4['time'] = pd.to_numeric(q4['time'])
q4['Order Date'] = pd.to_datetime(q4['Order Date'])
```

```python
q4.head()
```

Number of products sold in each month

```python
product_count = q4.groupby('month').count()
product_count
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(
    data=product_count.reset_index(),
    x = 'month',
    y = 'Product'
)
plt.show()
```

Amount of money brought in by month

```python
price_count = q4.groupby('month').sum()
price_count
```

```python
ax = sns.barplot(
    data = price_count.reset_index(),
    x = 'month',
    y = 'Price Each'
)
ax.set(xlabel='Month', ylabel='Income (Millions)')
plt.show()
```

Prices Of the different products

```python
product_price = q4[['Product','Price Each']].drop_duplicates(subset=['Product','Price Each'])
product_price
```

```python
sns.barplot(
    data=product_price,
    x = 'Product',
    y = 'Price Each'
)
plt.xticks(rotation=90)
plt.show()
```

Number of each product sold

```python
q4.groupby("Product").count()[['Order ID']]
```

```python
sns.barplot(
    data= q4.groupby("Product").count()[['Order ID']].reset_index(),
    x = 'Product',
    y = 'Order ID'
)
plt.xticks(rotation=90)
plt.show()
```

Sales of each product by city

```python
q4.groupby(['city','Product']).count()
```

```python
ax = sns.barplot(
    data=q4.groupby(['city','Product']).count().reset_index(),
    x = 'city',
    y = 'Order ID',
    hue = 'Product'
)
ax.set(xlabel='City', ylabel='Sales')
plt.show()
```

Sales of each product by state

```python
q4.groupby(['state','Product']).count()
```

```python
sns.barplot(
    data=q4.groupby(['state','Product']).count().reset_index(),
    x = 'state',
    y = 'Order ID',
    hue = 'Product'
)
```
